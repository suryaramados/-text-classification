```python
#train
import nltk.classify.util
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import names
 
def word_feats(s):
    return dict([(word, True) for word in s])
 

 
positive_features = [(word_feats(pos), 'pos') for pos in s]
negative_features = [(word_feats(neg), 'neg') for neg in s]
neutral_features = [(word_feats(neu), 'neu') for neu in s]
 
train_set = negative_features + positive_features + neutral_features
 
classifier = NaiveBayesClassifier.train(train_set) 
 
# Predict
neg = 0
pos = 0



for word in s:
    classResult = classifier.classify( word_feats(word))
    if classResult == 'neg':
        neg = neg + 1
    if classResult == 'pos':
        pos = pos + 1
 
print('Positive: ' + str(float(pos)/len(s)))
print('Negative: ' + str(float(neg)/len(s)))
```

```python
import re, math
from collections import Counter

WORD = re.compile(r'\w+')

def get_cosine(vec1, vec2):
     intersection = set(vec1.keys()) & set(vec2.keys())
     numerator = sum([vec1[x] * vec2[x] for x in intersection])

     sum1 = sum([vec1[x]**2 for x in vec1.keys()])
     sum2 = sum([vec2[x]**2 for x in vec2.keys()])
     denominator = math.sqrt(sum1) * math.sqrt(sum2)

     if not denominator:
        return 0.0
     else:
        return float(numerator) / denominator

def text_to_vector(text):
     words = WORD.findall(text)
     return Counter(words)

text1 = 'C:\\Users\\Surya\\Desktop\\quora(3)\\train1.csv'
text2 = 'C:\\Users\\Surya\\Desktop\\quora(3)\\test.csv'



vector1 = text_to_vector(text1)
vector2 = text_to_vector(text2)

cosine = get_cosine(vector1, vector2)

print('Cosine:', cosine)
```

```python
#The first pre-processing step which we will do is transform our tweets into lower case. 
#This avoids having multiple copies of the same words.
quoratrain['question_text'] = quoratrain['question_text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
quoratrain['question_text'].head()
```

```python
# Removing Punctuation
quoratrain['question_text'] = quoratrain['question_text'].str.replace('[^\w\s]','')
quoratrain['question_text'].head()
```

```python
#stop words (or commonly occurring words) should be removed from the text data
#Removal of Stop Words

from nltk.corpus import stopwords
stop = stopwords.words('english')
quoratrain['question_text'] = quoratrain['question_text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
quoratrain['question_text'].head()
```

```python
#Previously, we just removed commonly occurring words in a general sense. 
#We can also remove commonly occurring words from our text data First, 
#lets check the 10 most frequently occurring words in our text data then take call to remove or retain.
#Common word removal

freq = pd.Series(' '.join(quoratrain['question_text']).split()).value_counts()[:10]
freq
```

```python
#lets remove these words as their presence will not of any use in classification of our text data.

freq = list(freq.index)
quoratrain['question_text'] = quoratrain['question_text'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
quoratrain['question_text'].head()
```

```python
#Rare words removal

freq = pd.Series(' '.join(quoratrain['question_text']).split()).value_counts()[-10:]
freq
```

```python
All these above pre-processing steps are essential and
help us in reducing our vocabulary clutter so that the features produced in the end are more effective.
```
